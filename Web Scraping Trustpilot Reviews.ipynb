```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


data = []
# Set Trustpilot page numbers to scrape here
from_page = 1
to_page = 462

for i in range(from_page, to_page + 1):
    response = requests.get(f"https://uk.trustpilot.com/review/www.britishairways.com?page={i}")
    web_page = response.text
    soup = BeautifulSoup(web_page, "html.parser")

    for e in soup.select('article'):
        data.append({
            'Name': e.select_one('[data-consumer-name-typography]').text,
            'Review_Title':e.h2.text,
            'Review_date_original': e.select_one('[data-service-review-date-of-experience-typography]').text.split(': ')[-1],
            'Review_Rating':e.select_one('[data-service-review-rating] img').get('alt'),
            'Total_Reviews': e.select_one('[data-consumer-reviews-count-typography]').text,
            'Review_Text': e.select_one('[data-service-review-text-typography]').text if e.select_one('[data-service-review-text-typography]') else None,
            'Page_Number':i
        })
    # polite scraping
    time.sleep(1)

 # Convert to dataframe   
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Rating'] = [r.split(' ')[1] for r in df['Review_Rating']]
```

```python
df.drop(columns='Rating',inplace=True)
```

```python
# Save to a file
df.to_csv('BritishAirways_reviews.csv', index=False)
```
